```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl
%matplotlib inline
sns.set()
```

```python
titanic = pd.read_csv('https://raw.githubusercontent.com/daniel-dc-cd/data_science/master/module_3_Python/data/titanic.csv')

titanic["Survived"] = titanic["Survived"].replace([0],"Didn't Survived")
titanic["Survived"] = titanic["Survived"].replace([1],"Did Survived")
```

```python
# Age of the people in titanic 
titanic.Age.hist()
plt.title("Age")
plt.ylabel("The number of people")
plt.xlabel(" How old are they")
```

```python
# Count the number of Survived and number of did't
sns.countplot(x='Survived', data=titanic ,palette="Set2");
```

```python
# Count the number of Survived and number of did't of Male and Female
sns.catplot(
    data=titanic , x="Sex", col='Survived', kind='count', palette="Set2", height=6
)
```

```python
# Another simpler plot
fig_dims = (13, 7)
fig, ax = plt.subplots(figsize=fig_dims)
sns.countplot(x='Sex', data=titanic ,palette="Set2" , hue="Survived" ,ax=ax)
```

```python
# To see which Pclass have more people in it
fig_dims = (13, 7)
fig, ax = plt.subplots(figsize=fig_dims)
sns.countplot(x='Pclass', data=titanic ,palette="Set2" ,ax=ax)
```

```python
# Count the number of Survived and number of did't for each Pclass
sns.catplot(
    data=titanic , x="Pclass", col='Survived', kind='count', palette="Set2", height=6
)
```

```python
# Another simpler plot
fig_dims = (13, 7)
fig, ax = plt.subplots(figsize=fig_dims)
sns.countplot(x='Survived', data=titanic ,palette="Set2" , hue="Pclass" , ax=ax)
```

```python
# See the relation between Fare , Age and Pcalss
sns.lmplot(
    data=titanic,
    x="Age", y="Fare", hue="Pclass",
    height=6 , palette="muted"
)
```

```python
# The same relation of the figure above with another plot
fig_dims = (13, 7)
fig, ax = plt.subplots(figsize=fig_dims)
sns.scatterplot(x="Age", y="Fare", hue="Pclass",palette="muted", linewidth=0,data=titanic , ax=ax)
```

```python
sns.jointplot(x="Age", y="Fare", kind="hex", color="#4CB391" , data=titanic)
```

```python
# To see who survived or did't from each Pclass , age and the Fare in y axis
sns.relplot(x="Fare", y="Age", hue="Pclass", size="Survived",
            sizes=(150, 30), alpha=0.8, palette="muted",
            height=10, data=titanic)
```
