```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use('classic')
```

### Line plot :

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 20, 1000) # I Chose the from 0 to 20 to view it as more than 1 frequancy periods
plt.plot(x, np.sin(x) , color='blue',  linestyle='--' , label='sin(x)')
plt.plot(x, np.cos(x) , color='black' , linestyle=':' , label='cos(x)')
plt.xlim(0, 30)
plt.ylim(-1.2, 1.2);

# Or we can change the x and y axis limets like this :
#plt.axis([0, 25, -1.2, 1.2]);

plt.title("Sine and Cosine")
plt.xlabel("X")
plt.ylabel("sin(x) \n and \n cos(x)");
plt.legend();




```

### Scatter plot :

```python
plt.style.use('seaborn-whitegrid')

x = np.linspace(0, 20, 200)
y = np.sin(x)

plt.plot(x, y, 'o', color='blue' , linestyle='-');
# or we can do it like this :
#plt.plot(x, y, '-ob');
```

```python
# A second, more powerful method

plt.scatter(x, y, marker='o');
plt.xlim(0,25)
```

```python
#The primary difference of plt.scatter from plt.plot 
#is that it can be used to create scatter plots where the properties of each individual point 
#(size, face color, edge color, etc.) can be individually controlled or mapped to data.
```

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 800 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5,
            cmap='viridis')
plt.colorbar();  # show color scale
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T

plt.scatter(features[0], features[1], alpha=0.5,
            s=100*features[3], c=iris.target, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1]);
```

### Bar plot :

```python

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
Player = ['Robert Lewandowski', 'Sergio Aguero', 'Lionel Messi', 'Cristiano Ronaldo', 'Raheem Sterling']
Goals = [36,29,32,31,27]
ax.bar(Player , Goals , color=['red', 'blue', 'purple', 'green', 'lavender'])
plt.title("Goals in top five leagues in 2019")
plt.xlabel("Player Name")
plt.ylabel("Number of goals");


plt.show()
```
