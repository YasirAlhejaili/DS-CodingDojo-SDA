#### New data frame

#### Charts and tables 
Arabian gulf countries cases and deaths

```python
# Create colors
SA, UAE, KW, BA, QA, O = [plt.cm.Greens, plt.cm.RdPu, plt.cm.Blues , plt.cm.Reds , plt.cm.Oranges , plt.cm.Purples]

# Some calculations to use it in the charts
data1 = df_AG.groupby("countriesAndTerritories")["cases"].sum() # Total cases for each country

data2 = data1/df_AG.cases.sum()*100 # The cases percentage for each country from total cases in all the countries 

data3 = df_AG.groupby("countriesAndTerritories")["deaths"].sum() # Total deaths for each country

data4 = data3/data1*100 # The deaths percentage for each country from total cases in the same country
```

To see the total number of cases from each country

```python
t1 = pd.DataFrame(data1)
t1.sort_values(by=['cases'] , ascending=False)
```

To see the total number of deaths from each country

```python
t2 = pd.DataFrame(data3)
t2.sort_values(by=['deaths'] , ascending=False)
```

To merge the two tables

```python
t3 = t1.merge(t2, how='inner' , on="")
t3.sort_values(by=['cases'] , ascending=False)
```
